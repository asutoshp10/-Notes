```python
class linear_regression():
  def __init__(self):
    self.w=None
    self.b=None
    self.weights=None
    self.bias=None

  def fit(self,x,y,lr_rate=0.1,epochs=1):
    #x_tr=np.insert(x,0,1,axis=1)
    x_tr=x
    self.weights,self.bias=[],[]
    self.w=np.ones(x_tr.shape[1])
    self.b=np.zeros(1)
    for j in range(epochs):
      for i in range(x_tr.shape[0]):
        y_hat=np.dot(x_tr[i],self.w.T)+self.b
        self.w=self.w-lr_rate*-2*(y[i]-y_hat)*x_tr[i]
        self.b=self.b-lr_rate*-2*(y[i]-y_hat)
        self.weights.append(self.w)
        self.bias.append(self.b)
    return self.w,self.b

  def predict(self,x_test):
    y_pred=[]
    #x_test_=np.insert(x_test,0,1,axis=1)
    for i in range(x_test.shape[0]):
      y_pred.append(np.dot(x_test[i],self.w.T)+self.b)
    return np.array(y_pred)
```

```python
def my_train_test_split(x,y,test_size):
  n=[]
  y_test,x_test=[],[]
  for i in range(int(x.shape[0]*test_size)):
    a=np.random.randint(0,x.shape[0])
    n.append(a)
    y_test.append(y[a])
    x_test.append(x[a])
  x_tr=np.delete(x,np.array(n),axis=0)
  y_tr=np.delete(y,np.array(n),axis=0)
  return x_tr,np.array(x_test),y_tr,np.array(y_test)
```

```python
x_tr,x_test,y_tr,y_test=my_train_test_split(x,y,0.2)
```

```python
lr=linear_regression()
```

```python
lr.fit(x_tr,y_tr,0.01)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn import datasets
```

```python
x=np.linspace(1,10,300)
y=x+np.random.randn(x.shape[0])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x_tr,x_test,y_tr,y_test=train_test_split(x,y,test_size=0.2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_tr.reshape(-1,1),y_tr.reshape(-1,1))
```

```python
y_pred=lr.predict(x_test.reshape(-1,1))
```

```python
#preddicting the accuracy
r2_score(y_test.reshape(-1,1),y_pred)
```

```python
w=lr.coef_
b=lr.intercept_
print(w,b)
```

```python
x_l=np.linspace(1,10,300)
y_l=w.flatten()*x+b.flatten()
```

```python
plt.scatter(x,y)
plt.plot(x_l,y_l,color='red')
plt.show
```
